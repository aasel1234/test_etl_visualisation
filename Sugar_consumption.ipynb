```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

csv_file_stats = 'sugar_consumption_stats.csv'

data = pd.read_csv('g_sugar_per_person.csv')

summary_stats = (data.describe()).round(2)
print(summary_stats)

def load_to_csv(summary_stats):
    summary_stats.to_csv(csv_file_stats)
    
load_to_csv(summary_stats)

def visualise_plot(data):
    data_to_plot = data[['2012','2013','2014','2015','2016','2017','2018']].head(10)

    fig = plt.figure(figsize=(10,7))
    data_to_plot.boxplot()
    
    plt.title('Sugar Consumption (g per person) per Day')
    plt.ylabel('Sugar intake (g)')
    plt.xlabel('Year')
    plt.show()
    
visualise_plot(data)
```
